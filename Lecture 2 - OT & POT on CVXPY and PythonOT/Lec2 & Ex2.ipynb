## 1/ Recap Optimal Transport (OT) và Partial Optimal Transport
**Optimal Transport (OT)** là một thuật toán di chuyển tối ưu một khối lượng vật chất (mass) từ một phân phối $\mathbf{a}$ đến một phân phối $\mathbf{b}$.

Trong bài toán OT, chúng ta cần tìm lời giải tối ưu $\mathbf{X}$ cho bài toán tối ưu, với $\mathbf{X}$ là transport map biểu thị cách di chuyển tối ưu khối lượng giữa 2 phân phối : $$ \min_{\mathbf{X} \in U(\mathbf{a},\mathbf{b})} \langle \mathbf{C}^p, \mathbf{X} \rangle$$
với: 
+ $\sum_j \mathbf{X}_{i,j} = \delta_i^a$ (tổng cột tương đương với 1 phần tử trong $\mathbf{a}$)
+ $\sum_i \mathbf{X}_{i,j} = \delta_j^b$ (tổng hàng tương đương 1 phần tử trong $\mathbf{b}$)
+ $\mathbf{X} \geq 0$
+ $U(\mathbf{a}, \mathbf{b})$ là tập hợp các nghiệm $\mathbf{X}$ thỏa mãn 3 điều kiện trên: $$ U(a, b) = \{ X \mid \sum_i \sum_j \mathbf{X}_{i,j} = 1, \sum_j \mathbf{X}_{i,j} = \delta^a_i, \sum_i \mathbf{X}_{i,j} = \delta^b_j, \mathbf{X} \geq 0\},  $$

**Partial Optimal Transport (OT)** sẽ giãn điều kiện bắt buộc phải di chuyển hết các vật chất giữa phân phối $\mathbf{a}$ đến một phân phối $\mathbf{b}$.  
Nói cách khác, POT sẽ chỉ tập trung di chuyển một phần vật chất của phân phối $\mathbf{a}$ đến phân phối $\mathbf{b}$.  
Vì thế, bài toán POT được viết lại thành

## 2/ Sinkhorn algorithm

Thuật toán Sinkhorn là một thuật toán tính toán lời giải tối ưu cho bài toán optimal transport.  
Thuật toán Sinkhorn sử dụng entropic regularization term để giảm thời gian tính toán cũng như làm tăng tốc độ hội tụ.

### a) Wasserstein distance

Khoảng cách Wasserstein đo lường sự khác biệt giữa hai phân phối. Được tính bằng cách xác định nghiệm tối ưu $\mathbf{X}$ của bài toán OT.


Để đơn giản, chúng ta xét các phân phối rời rạc trên $\left[ \delta_1, \delta_2, \ldots, \delta_n \right]$.  
Cho 2 phân phối rời rạc là $\mathbf{a} = \left[ \delta_1^a, \delta_2^a, \ldots, \delta_n^a \right]$ và $\mathbf{b} = \left[ \delta_1^b, \delta_2^b, \ldots, \delta_n^b \right]$   
Ở đây, chúng ta sử dụng metric $L_2$ norm $c(x, y) = \|x - y\|_2$ để tính khoảng cách/cost để mang một mass từ điểm $x$ đến điểm $y$.  
Với khoảng cách $L_2$ norm, chúng ta có thể xây dựng một ma trận cost $\mathbf{C}$ là khoảng cách giữa 2 phân phối với: $$\mathbf{C}_{i,j} = c(\delta_i^a, \delta_i^b)$$

Trong khi đó, khoảng cách Wasserstein-$p$ giữa phân phối $\mathbf{a}$ và $\mathbf{b}$ được định nghĩa như sau

$$ 
W_p(a, b) = \left( \min_{\mathbf{X} \in U(a,b)} \langle \mathbf{C}^p, \mathbf{X} \rangle \right)^{1/p} = \left( \min_{\mathbf{X} \in U(\mathbf{a},\mathbf{b})} \sum_{i,j} \mathbf{C}_{i,j}^p \mathbf{X}_{i,j} \right)^{1/p}, 
$$

trong đó: $U(\mathbf{a}, \mathbf{b})$ là tập hợp các nghiệm $\mathbf{X}$ thỏa mãn 3 điều kiện trên: $$ U(a, b) = \{ X \mid \sum_i \sum_j \mathbf{X}_{i,j} = 1, \sum_j \mathbf{X}_{i,j} = \delta^a_i, \sum_i \mathbf{X}_{i,j} = \delta^b_j, \mathbf{X} \geq 0\},  $$

là phân phối chung trên: $\mathbf{a} \times \mathbf{b}$:
$$
[\delta_1^a, \delta_2^a, \ldots, \delta_n^a] \times [\delta_1^b, \delta_2^b, \ldots, \delta_n^b].
$$

```python
# Ví dụ của khoảng cách Wasserstein của 2 phân phối

# Khởi tạo phân phối a là mixed Gaussian N(1, 2) và N(10, 1.5), và phân phối b là Gamma(7, 1)
n_bins = 50
# Helper function to sample a mixed Gaussian distribution
def mixedGaussian(mu1, mu2, sigma1, sigma2, n):
    bernoulli = np.random.binomial(n = 1, p = 0.5, size = n)
    gaussian1 = np.random.normal(mu1, sigma1, n)
    gaussian2 = np.random.normal(mu2, sigma2, n)
    return (gaussian1**bernoulli)*(gaussian2**(1-bernoulli))


# Lấy các mẫu từ phân phối mixed Gaussian distribution N(1, 2) và N(10, 1.5)
dist_a = mixedGaussian(mu1 = 1, mu2 = 10, sigma1 = 2, sigma2 = 1.5, n = 100000)
p_a, edges_a = np.histogram(dist_a, bins=n_bins)
p_a = p_a/100000


# Lấy các mẫu từ phân phối from Gamma(7, 1)
dist_b = np.random.gamma(7, scale = 1, size = 100000)
p_b, edges_b = np.histogram(dist_b, bins=n_bins)
p_b = p_b/100000


# Vẽ 2 phân phối
pa = figure(title='dist_a: mixed gaussian distribution, 0.5*N(1, 2) + 0.5*N(10, 1.5)',
            background_fill_color="#fafafa", tools="save", height=600)
pa.quad(top=p_a, bottom=0, left=edges_a[:-1], right=edges_a[1:],
        fill_color="navy", line_color="white", alpha=0.5)
pb = figure(title='dist_b: gamma distribution, Gamma(7, 1)',
             background_fill_color="#fafafa", y_range = pa.y_range, height=600)
pb.quad(top=p_b, bottom=0, left=edges_b[:-1], right=edges_b[1:], fill_color="navy", line_color="white", alpha=0.5)
show(row(pa, pb))
```

```python
# Sử dụng Linear Programming để tính khoảng cách Wasserstein
edges_a = edges_a[:-1]
edges_b = edges_b[:-1]
# Ma trận Cost sử dụng L2 norm
C = (edges_a.reshape((n_bins,1)) - edges_b.reshape((1,n_bins)))**2
# Tạo biến cho bài toán LP, sử dụng nonneg để giữ X >= 0
P0 = cp.Variable((n_bins, n_bins), nonneg=True)

# Thêm 2 điều kiện ràng buộc: tổng hàng bằng phân phối b và tổng cột bằng phân phối a.
constraints = [cp.sum(P0, axis = 1) == p_a, cp.sum(P0, axis = 0) == p_b]

# Tạo hàm mục tiêu
obj = cp.Minimize(cp.trace(C.T@P0))

# Giải bài toán.
prob = cp.Problem(obj, constraints)
prob.solve()  # Returns the optimal value.
print("Khoảng cách Wasserstein", prob.value)
print("Nghiệm bài toán", P0.value)
```

### b) Entropic regularization (Entropic penalized Wasserstein distance)

#### b.1) KL divergence - Độ lệch (Entropy) tương đối

Kullback-Leibler (KL) divergence (độ lệch KL) là một thang đo sự giống nhau của 2 phân phối.  
Trong sác xuất thống kê và ML, KL-divergence được dùng để đo độ sai lệch của phân phối đã học được bởi các model so với phân phối chuẩn của dữ liệu.

**Định Nghĩa**
Cho 2 phân phối $\mathbf{P}$ và $\mathbf{Q}$, KL divergence giữa $\mathbf{Q}$ và $\mathbf{P}$ được tính như sau:

$$
KL(P \mid \mid Q) = \sum_{x \in X} P(x) \log \frac{P(x)}{Q(x)}
$$

Từ định nghĩa của KL divergence, độ lệch tương đối giữa $\mathbf{X}$ và $\mathbf{a} \otimes \mathbf{b}$ như sau

\begin{align}
KL(\mathbf{X} \mid \mathbf{a} \otimes \mathbf{b}) &= \sum_{i,j} \mathbf{X}_{i,j} \log \frac{\mathbf{X}_{i,j}}{a_i \times b_j} \\
&= \sum_{i,j} \mathbf{X}_{i,j} \log \mathbf{X}_{i,j} - \sum_{i,j} \mathbf{X}_{i,j} \log a_i \times b_j \\
&= \sum_{i,j} \mathbf{X}_{i,j} \log \mathbf{X}_{i,j} - \sum_{i} a_i \log a_i - \sum_{j} b_j \log b_j,
\end{align}

với $\mathbf{a} \otimes \mathbf{b}$ là phân phối chung của hai phân phối biên $\mathbf{a}$ và $\mathbf{b}$ độc lập.  

Theo phép biến đổi trên, thuật ngữ regularization $KL(\mathbf{X} \mid \mathbf{a} \otimes \mathbf{b})$ tương đương với

$$
-H(\mathbf{X}) = \sum_{i,j} \mathbf{X}_{i,j} \log \mathbf{X}_{i,j} - \mathbf{X}_{i,j}
$$

vì sự khác biệt là một hằng số không liên quan đến **P**.  

#### b.2) Entropic regularization (Entropic penalized Wasserstein distance)

Trong bài toán OT và POT, một hàm entropy $H(\mathbf{X})$ thường được thêm vào hàm cost để tăng tốc độ hội tụ và giảm thời gian chạy của thuật toán:
$$\mathbf{H(X)} = -\sum^n_{i=1}log(P(\mathbf{X}_i))P(\mathbf{X}_i) = - \langle P(\mathbf{X}), log(P\mathbf{X}) \rangle - \mathbb{E}_\mathbf{X} \left[ log(P(\mathbf{X})) \right]$$

Hàm entropy này có thể viết dưới dạng KL, và vì thế ta có thể viết lại bài toán tính Wasserstein distance với hàm entropy là $KL(\mathbf{X} | \mathbf{a} \otimes \mathbf{b})$:
\begin{align}
W_{p, \epsilon}^p &= \min_{\mathbf{X} \in U(\mathbf{a},\mathbf{b})} \langle C^p, \mathbf{X} \rangle + \epsilon KL(\mathbf{X} \mid \mathbf{a} \otimes \mathbf{b}) \\
 &= \min_{\mathbf{X} \in U(\mathbf{a},\mathbf{b})} \langle C^p, \mathbf{X} \rangle - \epsilon H(\mathbf{X}) \\
 &= \min_{\mathbf{X} \in U(\mathbf{a},\mathbf{b})} \langle C^p, \mathbf{X} \rangle + \epsilon \left(\sum_{i,j} \mathbf{X}_{i,j} \log \mathbf{X}_{i,j} - \mathbf{X}_{i,j}\right)
\end{align}
Với $\epsilon$ là regularization coefficient.

Việc thêm vào một entropic regularization có tác dụng:
+ Giúp cho Wasserstein distance có một nghiệm duy nhất, khi $KL(\mathbf{X} | \mathbf{a} \otimes \mathbf{b})$ là một hàm lồi mạnh (strongly-convex)
+ Nghiệm $\mathbf{X}$ trả ra sẽ có nhiều phần tử khác 0 hơn (giảm sai số và những vấn đề thao tác khác)  
+ Khi $\epsilon \rightarrow \infty$, nghiệm $\mathbf{X} \rightarrow a \otimes b$; khi $\epsilon \rightarrow 0$, nghiệm $\mathbf{X} \rightarrow P^{OT}$.

#### b.3) Công thức dual formation trong bài toán tối ưu

Trong bài toán tối ưu hóa, công thức đối ngẫu (dual formation) là một cách biểu diễn thay thế cho một bài toán tối ưu hóa.  
Nó cung cấp một góc nhìn khác giúp giải quyết hoặc phân tích một số loại bài toán dễ dàng hơn.  

Cho một bài toán tối ưu nguyên thủy:
\begin{align}
   & \min_{x} && f(x) \\
   & \text{subject to} && g_i(x) \leq 0, \quad i = 1, \ldots, m \\
   & && h_j(x) = 0, \quad j = 1, \ldots, p
\end{align}

Với bài toán dual formation, chúng ta sẽ thêm vào bài toán các dual variable (Lagrange multiplier) vào các ràng buộc đẳng thức (equality constraint) và ràng buộc bất đẳng thức (inequality constraint)
$$ L(x, \lambda, \nu) = f(x) + \sum_{i=1}^m \lambda_i g_i(x) + \sum_{j=1}^p \nu_j h_j(x) $$

Lúc này, ta sẽ nghịch đảo bài toán (minimize thành maximize và ngược lại). Đồng thời, hàm mục tiêu sẽ thay đổi từ $f(x)$ thành:
$$ g(\lambda, \nu) = \inf_{x} L(x, \lambda, \nu) $$

Và bài toán trở thành:
\begin{align}
   & \max_{\lambda, \nu} && g(\lambda, \nu) \\
   & \text{subject to} && \lambda_i \geq 0, \quad i = 1, \ldots, m
\end{align}

Điều kiện bổ sung (Complementary Slackness)
Điều kiện bổ sung cung cấp mối quan hệ giữa nghiệm nguyên thủy và nghiệm dual variable.  
Nó cho biết rằng với mỗi $i$:
- Nếu $\lambda_i > 0$, thì $g_i(x) = 0$.
- Nếu $g_i(x) < 0$, thì $\lambda_i = 0$.

Ví dụ:
+ Bài toán nguyên thủy (primal problem)
\begin{aligned}
& \min_{x} && c^T x \\
& \text{subject to} && Ax \geq b \\
& && x \geq 0
\end{aligned}

+ Bài toán đối ngẫu (dual problem)
\begin{aligned}
& \max_{\lambda} && b^T \lambda \\
& \text{subject to} && A^T \lambda \leq c \\
& && \lambda \geq 0
\end{aligned}

Ở đây, $\lambda$ là các biến đối ngẫu tương ứng với ràng buộc $Ax \geq b $.

**Lợi ích của lý thuyết đối ngẫu**
- Weak Duality: Giá trị tối ưu của bài toán đối ngẫu cung cấp một cận dưới cho giá trị tối ưu của bài toán nguyên thủy.
- Strong Duality: Dưới một số điều kiện các giá trị tối ưu của bài toán nguyên thủy và bài toán đối ngẫu là bằng nhau.
- Sensitivity Analysis: Các biến đối ngẫu cung cấp cái nhìn sâu sắc về cách sự thay đổi trong các ràng buộc ảnh hưởng đến nghiệm tối ưu của bài toán nguyên thủy.

**Tóm tắt** 
- Công thức dual formation: Một cách biểu diễn thay thế cho một bài toán tối ưu hóa bằng cách giới thiệu các biến đối ngẫu liên quan đến các ràng buộc.
- Dual Variable (Lagrange multiplier): Các biến được giới thiệu để thực thi các ràng buộc trong bài toán nguyên thủy và cung cấp cái nhìn sâu sắc về độ nhạy của bài toán.
- Điều kiện bổ sung (Complementary Slackness): Một điều kiện liên kết các nghiệm nguyên thủy và đối ngẫu, giúp xác định tính tối ưu.

```python
# Tạo random các dữ kiện bài toán (cho phù hợp với bài toán quy hoạch tuyến tính)
m = 15
n = 10
np.random.seed(1)
s0 = np.random.randn(m)
lamb0 = np.maximum(-s0, 0)
s0 = np.maximum(s0, 0)
x0 = np.random.randn(n)
A = np.random.randn(m, n)
b = A @ x0 + s0
c = -A.T @ lamb0
```

```python
# Tạo bài toán trên CVXPY và giải bài toán
x = cp.Variable(n)  # Khai báo biến x
# Cho bài toán và các điều kiện ràng buộc vào CVXPY
objective = cp.Minimize(c.T@x)  # Hàm mục tiêu
constraints = [A @ x <= b]  # Các điều kiện ràng buộc
prob = cp.Problem(objective,
                  constraints)
prob.solve()  # Giải bài toán

# Print result.s
print("\nGiá trị tối ưu là", prob.value)
print("Nghiệm x tối ưu là")
print(x.value)

# Lấy kết quả của dual formulation
print("Nghiệm dual solution")
print(prob.constraints[0].dual_value)
```

#### b.4) Sinkhorn algorithm

Trong phần sau, chúng ta xét khoảng cách Wasserstein-2, cụ thể là $p = 2$.  
Sau đó, chúng ta viết lại $\mathbf{C}^2$ thành $\mathbf{C}$. 

Thuật toán Sinkhorn sử dụng công thức đối ngẫu của bài toán tối ưu lồi bị ràng buộc, chuyển các biến chưa biết từ **P** ($n^2$ biến chưa biết) thành các biến đối ngẫu $\mathbf{f}$, $\mathbf{g}$ ($2n$ biến chưa biết) của các ràng buộc tuyến tính.

Định nghĩa hàm Lagrange là

$$
L(\mathbf{P}, \mathbf{f}, \mathbf{g}) = \langle \mathbf{C}, \mathbf{P} \rangle - \epsilon H(\mathbf{P}) - \langle \mathbf{f}, \mathbf{P1} - \mathbf{a} \rangle - \langle \mathbf{g}, \mathbf{P}^T\mathbf{1} - \mathbf{b} \rangle.
$$

Điều kiện bậc nhất là

$$
\frac{\partial L(\mathbf{P}, \mathbf{f}, \mathbf{g})}{\partial P_{i,j}} = C_{i,j} + \epsilon \log P_{i,j} - f_i - g_j = 0,
$$

dẫn đến nghiệm

$$
\mathbf{P} = \text{diag}(e^{\mathbf{f} / \epsilon}) * e^{- \mathbf{C} / \epsilon} * \text{diag}(e^{\mathbf{g} / \epsilon}).
$$

Do đó, nghiệm phải ở dạng

$$
\text{diag}(\mathbf{u}) * \mathbf{K} * \text{diag}(\mathbf{v}), \mathbf{K} = e^{- \mathbf{C} / \epsilon}.
$$

Các ràng buộc biên yêu cầu

$$
\text{diag}(\mathbf{u}) * \mathbf{K} * \text{diag}(\mathbf{v}) \mathbf{1} = \mathbf{a} \text{ và } \text{diag}(\mathbf{v}) * \mathbf{K}^T * \text{diag}(\mathbf{u}) \mathbf{1} = \mathbf{b}.
$$

**Đầu vào:** $\mathbf{C}, \mathbf{a}, \mathbf{b}, \epsilon$

- Khởi tạo: $\mathbf{u} = \mathbf{v} = 1, \mathbf{K} = e^{- \mathbf{C} / \epsilon}$.
- Vòng lặp chính:
  - Trong khi $\|\mathbf{P}\|$ thay đổi, thực hiện:
    $$
    \mathbf{u}^{(i+1)} = \frac{\mathbf{a}}{\mathbf{K}\mathbf{v}^{(i)}}, \mathbf{v}^{(i+1)} = \frac{\mathbf{b}}{\mathbf{K}^T\mathbf{u}^{(i+1)}}.
    $$

**Trả về:** $\mathbf{P} = \mathbf{u} * \mathbf{K} * \mathbf{v}, \hat{W}_p^p = \text{trace}(\mathbf{C}^T\mathbf{P})$


```python
# Sinkhorn làm tay
element_max = np.vectorize(max)
def sinkhorn(C, a, b, epsilon, precision):
    a = a.reshape((C.shape[0], 1))
    b = b.reshape((C.shape[1], 1))
    K = np.exp(-C/epsilon)
    
    # initialization
    u = np.ones((C.shape[0], 1))
    v = np.ones((C.shape[1], 1))
    P = np.diag(u.flatten()) @ K @ np.diag(v.flatten())
    p_norm = np.trace(P.T @ P)
    
    while True:
        u = a/element_max((K @ v), 1e-300) # avoid divided by zero
        v = b/element_max((K.T @ u), 1e-300)
        P = np.diag(u.flatten()) @ K @ np.diag(v.flatten())
        if abs((np.trace(P.T @ P) - p_norm)/p_norm) < precision:
            break
        p_norm = np.trace(P.T @ P)
    return P, np.trace(C.T @ P)
```

```python
###       apply sinkhorn algorithm to epsilon = 0.1, 1, 10, 100       ###
P1, W_1 = sinkhorn(C, p_a, p_b, epsilon=1, precision=1e-30)
P01, W_01 = sinkhorn(C, p_a, p_b, epsilon=0.1, precision=1e-30)
P10, W_10 = sinkhorn(C, p_a, p_b, epsilon=10, precision=1e-30)
P100, W_100 = sinkhorn(C, p_a, p_b, epsilon=100, precision=1e-30)
```

### Giải bài toán Optimal Transport bằng thư viện PythonOT


Có 8 tiệm bánh và 5 tiệm cafe trong thành phố.  
Các tiệm bánh sản xuất lần lượt là: 31, 48, 82, 30, 40, 48, 89, 73  
Các tiệm cafe lần lượt bán số bánh mỗi ngày là: 82, 88, 92, 88, 91  
Bài toán: Làm sao cung cấp đủ số bánh cho các quán cafe, đồng thời giảm tối thiểu giá vận chuyển

Hoàn thành formulation và giải bài toán bằng CVXPY

**Formulation**:
$$
\begin{split}
\text{minimize} & \,\,\, \sum_{i=1}^m \sum_{j=1}^n c_{ij} x_{ij} \\
\text{subject to} & \,\,\,
\begin{split}
\sum_{j=1}^n x_{ij} &= s_i \quad \forall i \\
\sum_{i=1}^m x_{ij} &= d_j \quad \forall j \\
x_{ij} &\geq 0
\end{split}
\end{split}
$$

Với:
+ $c_{ij}$: chi phí vận chuyển từ tiệm bánh i đến tiệm cafe j
+ $x_{ij}$: số lượng bánh vận chuyển từ tiệm bánh i đến tiệm cafe j
+ $s_i$: số bánh từ tiệm bánh i
+ $d_j$: số bánh bán từ tiệm cafe j

```python
# Tạo cost matrix
cost_matrix = np.zeros((len(bakery_pos), len(cafe_pos)))
for i in range(len(bakery_pos)):
    dist_to_cafe = []
    for j in range(len(cafe_pos)):
        cost_matrix[i, j] = (np.linalg.norm(bakery_pos[i] - cafe_pos[j]))
cost_matrix

# Quy định biến, quy định integer vì số bánh là số nguyên
x = cp.Variable((len(bakery_pos), len(cafe_pos)), integer=True)

# Quy định hàm mục tiêu
objective = cp.Minimize(cp.sum(cp.multiply(cost_matrix, x)))

# Thêm các điều kiện ràng buộcc
constraints = [
    # Đảm bảo số bánh đến các quán cafe không quá số lượng bánh ở mỗi tiệm bánh
    cp.sum(x, axis=1) <= bakery_prod,
    # Đảm bảo các quán cafe đủ số bánh bán ra
    cp.sum(x, axis=0) >= cafe_prod,
    x >= 0                              # Đảm bảo tất cả các biến >= 0
]

# Giải bài toán tối ưu
problem = cp.Problem(objective, constraints)
optimal = problem.solve()

# In ra kết quả
print("Chi phí vận chuyển tối ưu:", optimal)
print("Đường đi tối ưu:")
for i in range(len(x.flatten().value)):
    if x.flatten().value[i] > 0:
        print('Tiệm bánh ' + label_list[i][1:2] + ' đến quán cafe ' +
              label_list[i][4:] + ' ' + str(int(x.flatten().value[i])) + ' bánh')

# vẽ các tiệm cafe và bánh trên bản đồ.
# Kích cỡ của điểm tròn biểu thị sản lượng bánh được làm/bán
# Độ đậm nhạt của đường thẳng biểu thị số bánh được giao
plt.figure()
plt.imshow(data['Imap'], interpolation='bilinear')
plt.scatter(bakery_pos[:, 0], bakery_pos[:, 1], color='b', s=bakery_prod)
plt.scatter(cafe_pos[:, 0], cafe_pos[:, 1], color='r', s=cafe_prod)
for i in range(len(bakery_pos)):
    for j in range(len(cafe_pos)):
        plt.plot([bakery_pos[i, 0], cafe_pos[j, 0]], [bakery_pos[i, 1], cafe_pos[j, 1]],
                 '-k', lw=3. * x.value[i, j] / x.value.max())
# Plot bakery positions with labels
for i, pos in enumerate(bakery_pos):
    plt.text(pos[0], pos[1] - 10, f'b{i+1}', color='b', ha='center')

# Plot cafe positions with labels
for i, pos in enumerate(cafe_pos):
    plt.text(pos[0], pos[1] - 10, f'c{i+1}', color='r', ha='center')

plt.legend(['Bakery', 'Cafe'])
plt.title('Manhattan Bakeries and Cafés')
plt.show()
```
